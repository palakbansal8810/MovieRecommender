```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
movies.shape
```

```python
movies=movies.merge(credits,on='title')
movies.shape
```

```python
movies.original_language.unique()
```

```python
movies.head(1).columns
```

```python
# genres,id,keywords,title,overview,cast,crew
```

```python
movies=movies[['genres','id','keywords','title','overview','cast','crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convt(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies.genres.apply(convt)
movies.head()
```

```python
movies.iloc[1].keywords
```

```python

def keywords(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
keywords(movies.iloc[1].keywords)
```

```python
movies['keywords']=movies.keywords.apply(keywords)
```

```python
movies.head()
```

```python
movies.iloc[1].cast
```

```python
def cast(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L[0:3]
```

```python
cast(movies.iloc[1].cast)
```

```python
movies['cast']=movies.cast.apply(cast)
```

```python
movies.head()
```

```python
movies.cast.value_counts
```

```python
movies.iloc[1].crew
```

```python
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
director(movies.iloc[1].crew)
```

```python
movies['crew']=movies.crew.apply(director)
```

```python
movies.head()
```

```python
movies['overview']=movies.overview.apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies.genres.apply(lambda x: [i.replace(" ",'') for i in x])
movies['crew']=movies.crew.apply(lambda x: [i.replace(" ",'') for i in x])
movies['cast']=movies.cast.apply(lambda x: [i.replace(" ",'') for i in x])

movies['keywords']=movies.keywords.apply(lambda x: [i.replace(" ",'') for i in x])
```

```python
movies.head(1)
```

```python
movies['tags']= movies.crew + movies.cast + movies.genres+ movies.overview + movies.keywords
movies.head()
```

```python
new_df=movies[['id','title','tags']]
new_df.head(1)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.iloc[0].tags
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head(1)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
new_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x : x[1])[0:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
new_df[new_df['title']=='Batman'].index[0]
```

```python
new_df.iloc[40].title
```

```python
recommend('Cars 2')
```
